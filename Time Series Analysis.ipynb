# Time Series Analysis and Forecasting (using ARIMA):


```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Generate synthetic time series data
np.random.seed(0)
time_steps = np.arange(0, 100)
true_values = 0.5 * time_steps + np.sin(0.2 * time_steps) + np.random.normal(0, 1, size=len(time_steps))
```

```python
# Fit ARIMA model
model = ARIMA(true_values, order=(1, 1, 1))
results = model.fit()
```

```python
# Forecast future values
forecast_steps = 30
forecast = results.forecast(steps=forecast_steps)
```

```python
# Plot original data and forecast
plt.plot(time_steps, true_values, label='Original Data')
plt.plot(np.arange(len(true_values), len(true_values) + forecast_steps), forecast, label='Forecast', linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
